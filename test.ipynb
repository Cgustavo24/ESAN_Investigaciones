```python
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

# Ruta de la imagen
imagen_path = 'piura_20170227t234245_x-3-_y-40.png'

# Cargar la imagen
img = Image.open(imagen_path)

# Obtener los puntos del polgono
puntos = [[176.93617021276597, 64.59574468085107],
          [190.9787234042553, 81.19148936170212],
          [197.36170212765958, 102.46808510638297],
          [205.4468085106383, 97.7872340425532],
          [211.40425531914892, 108.0],
          [203.3191489361702, 110.97872340425532],
          [196.51063829787236, 114.38297872340425],
          [189.27659574468086, 102.04255319148936],
          [182.04255319148936, 92.68085106382979],
          [175.23404255319147, 86.29787234042553]]

# Crear la figura y el eje
fig, ax = plt.subplots()

# Mostrar la imagen
ax.imshow(img)

# Crear el polgono con los puntos
poligono = Polygon(puntos, closed=True, edgecolor='r', linewidth=2, fill=None)

# Aadir el polgono al eje
ax.add_patch(poligono)

# Ajustar los lmites del grfico
ax.set_xlim(0, img.width)
ax.set_ylim(img.height, 0)  # Invertir el eje y para que coincida con el sistema de coordenadas de la imagen

# Mostrar el grfico

plt.title('Polgono sobre la imagen')
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
